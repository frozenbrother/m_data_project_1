```python
# Declare dependencies
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
import os as os

#Display data summary
crash_data = os.path.join("resources/Crash_Data.csv")
crash_data = pd.read_csv(crash_data, low_memory=False)
crash_data.head()
```

```python
# Create bar chart displaying types of road users involved in fatal accidents

#my_xticks = ['Motorcycle Rider', 'Pedestrian', 'Car Passenger', 'Car Driver', 'Other', 'Cyclist', 'Motorcycle passenger']

plt.bar(crash_data['Road User'].unique(),
        crash_data['Road User'].value_counts(), 
        color='crimson')
plt.rcParams["figure.figsize"] = (28, 10)
plt.xlabel('Types of Road Users', fontdict = {'fontsize' : 20})
plt.ylabel('Total Fatal Accidents (1989-2011)', fontdict = {'fontsize' : 20})
plt.title('Road Users Involved in Fatal Accidents', fontdict = {'fontsize' : 50})
#ax.set_xticklabels = 'Motorcycle Rider', 'Pedestrian', 'Car Passenger', 'Car Driver', 'Other', 'Cyclist', 'Motorcycle passenger'

#plt.xticks(x, my_xticks)
#plt.show()
plt.savefig('Road Users Bar Chart.png')
```
