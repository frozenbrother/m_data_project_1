```python
# Declare dependencies
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
import os as os

#Display data summary
crash_data = os.path.join("resources/Crash_Data.csv")
crash_data = pd.read_csv(crash_data, low_memory=False)
crash_data.head()
```

```python
# Create bar chart displaying types of road users involved in fatal accidents

plt.bar(crash_data['Road User'].unique(),
        crash_data['Road User'].value_counts(), 
        color='crimson')
plt.rcParams["figure.figsize"] = (28, 10)
plt.ylabel('Total Fatal Accidents (1989-2011)', fontdict = {'fontsize' : 30})
plt.xticks([0, 1, 2, 3, 4, 5, 6], ['Motorcycle Rider', 'Pedestrian', 'Car Passenger', 'Car Driver', 'Other', 'Cyclist', 'Motorcycle passenger'], fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Road Users Involved in Fatal Accidents', fontdict = {'fontsize' : 50})

# Save plot as PNG
plt.savefig('Road Users Bar Chart.png')

# Display Plot
plt.show()
```

```python
# Create bar chart displaying types of road users involved in fatal accidents
crash_pie = crash_data['Road User'].value_counts()
my_labels = 'Motorcycle Rider', 'Pedestrian', 'Car Passenger', 'Car Driver', 'Other', 'Cyclist', 'Motorcycle passenger'
myexplode = [0.1, 0, 0, 0, 0, 0, 0]
plt.pie(crash_pie, 
        labels=my_labels, 
        startangle=0, 
        #autopct='%1.1f%%', 
        explode = myexplode,
        shadow=True,
        textprops={'fontsize': 10})
plt.title('Road Users Involved in Fatal Accidents', fontdict = {'fontsize' : 30})
plt.axis('equal')

# Save plot as PNG
#plt.savefig('Road Users Pie Chart.png')

# Display Plot
#plt.show()
```
